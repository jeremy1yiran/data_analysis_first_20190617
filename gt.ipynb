```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

# 构造正态分布的概率密度函数
x = np.arange(-5, 5, 0.1)
print (x)
y = stats.norm.cdf(x, 0, 1)
plt.plot(x, y)
```

```python
# 造一个正态分布的概率密度函数

import numpy as np
import matplotlib.pyplot as plt
data = np.random.normal(loc=7,scale=1,size=100) #均值mean,标准差std,数量
#print(len(data))
#print(np.arange(len(data)))  #数组小标打印
sorted_ = np.sort(data) #排序
#print(sorted_)
#print(float(len(sorted_)))
yvals = np.arange(len(sorted_))/float(len(sorted_))
print(yvals)
plt.plot(sorted_, yvals)

```

```python
x_label = stats.norm.ppf(yvals)  #分位点函数（CDF的逆）
print (x_label); 
plt.scatter(x_label, sorted_)
```

既然做对比那么对比的标准是什么呢，我们说如果所有点基本上在一条直线上，我们
可以说这两个分布是同一分布。所以根据qq图，我们得出结论目标数据组服从正太分#布。

```python
stats.probplot(data, dist="norm", plot=plt)
plt.show()
```
